```python
### linux  print all secondary path to a file Res_ctrl 
###  ls -R | awk '{print i$0} i=`pwd`'/' | sed ''
import json
import os
import sys
output=[]
raw_gz_vcf=[]

pwd='/mnt/datashare/home/wangzesong/HAPE/MT/MT_DNA_seqencing/Res_ctrl/'
i=0
#a="/mnt/datashare/home/wangzesong/HAPE/MT/MT_DNA_seqencing/Res_ctrl/./08H:"
#a.(line.split('/')[-1]).split(':')[-2]

## get sample name 
with open('res_ctrl_path.txt') as res_ctrl_path:
   for line in res_ctrl_path:
        if ":" in line :
           output.append((line.split('/')[-1]).split(':')[-2])
        elif "vcf.gz" in line :
            dict_file={}
            #path_list=[]   
            dict_file["class"] = 'File'
            dict_file["path"] = pwd + output[i] + "/" + line.split('/')[-1][:-1]
            #path_list.append(dict_file)
            raw_gz_vcf.append(dict_file)
            i=i+1
#output
#raw_gz_vcf
for i in range(6):
   sample=[]
   input1 = []
   dict3 = {}
   dict4={}
   dict4["class"]="Directory"
   dict4['path']="/mnt/datashare/annovar_db/humandb"
   dict3['annovar_db']=dict4
   for k in range(10):
      sample.append(output[k+i*10])
      input1.append(raw_gz_vcf[k+i*10])
      dict3['output']=sample
      dict3['vcf_gz']=input1
      string = json.dumps(dict3,indent =2)
      with open('Res_ctrl_param_'+str(i)+'.json','w') as json_res:
         json_res.write(string)
```
