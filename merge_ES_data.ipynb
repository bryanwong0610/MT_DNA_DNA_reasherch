```python
from fileinput import filename

from pandas import read_csv

with open('sample_info.info',encoding='utf-8') as sample_info:
   lines=sample_info.readlines()
line1=lines[0].split(',')
dt_dict={}
for i in line1:
   file_name=i+'_Final_variation_filterd.csv'
   b=str(file_name)
   dt=read_csv("./"+b)

   a=[]
   file_lenth=len(dt)
   for le in range(file_lenth):
      a.append(i)
   dt[i]=a
   dt['InheritanceModes']=dt['InheritanceModes'].astype('str')
   dt['exomiserRank']=dt['exomiserRank'].astype('str')
   dt['combinedScore']=dt['combinedScore'].astype('str')
   dt['Genotype']=dt['Genotype'].astype('str')
   dt['Hom_Het']=dt['Hom_Het'].astype('str')
   dt['Depth']=dt['Depth'].astype('str')
   dt['Allele_Depth']=dt['Allele_Depth'].astype('str')
   dt['WT_ratio']=dt['WT_ratio'].astype('str')
   dt[i]=dt[i]+ ': Iheritance:'+dt['InheritanceModes'] +'. ExomiserRank:'+ dt['exomiserRank'] +'. ExomiserScore:'+  dt['combinedScore'] +'. Genotype:'  +  dt['Genotype'] +'. Hom_Het:' + dt['Hom_Het'] + '. Depth:'+dt['Depth'] +'. Allele_Depth:'+  dt['Allele_Depth'] + '. WT_ratio'+ dt['WT_ratio']
   del dt['Allele_Depth'] 
   del dt['InheritanceModes']
   del dt['exomiserRank']
   del dt['combinedScore']
   del dt['Genotype']
   del dt['Hom_Het']
   del dt['Depth']
   del dt['WT_ratio']
   dt_dict[i]=dt
raw_dt=dt_dict['GF-E-18']
line2=line1[1:]
merge_rule=raw_dt.columns.values.tolist()
merge_rule=merge_rule[:-1]
reshape_col=line1+merge_rule
for x in line2:
   mer_dt=pd.merge(raw_dt,dt_dict[x],how='outer',on=merge_rule)
   raw_dt=mer_dt
raw_dt=raw_dt[reshape_col]
raw_dt.to_csv('HAPE_merge_filterd.csv')
```
