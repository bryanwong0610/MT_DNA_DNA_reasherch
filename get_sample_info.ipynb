```python
from operator import index
from os import sep
from random import sample
import pandas as pd
import numpy as py 

result_info = pd.read_csv('Result_info.txt',sep='\t')
res_name = result_info['res_name']
sample_name = result_info['sample_name']
dt_dict={}
a=[]
i=0
for line in sample_name :
    file_nm= line +'.hg19_multianno.txt'
    file = pd.read_table(file_nm,sep='\t')
    a.append(line)
    sam_name=sample_name[i]
    file[sam_name] = sam_name
    i=i+1
    dt_dict[line]=file
raw_dt=dt_dict['08H-1']
merge_rule=raw_dt.columns.values.tolist()
merge_rule=merge_rule[:-1]
sampe_after=sample_name[1:]
reshape_col= a + merge_rule
for x in sampe_after :
   mer_dt=pd.merge(raw_dt,dt_dict[x],how='outer',on=merge_rule)
   raw_dt=mer_dt
raw_dt=raw_dt[reshape_col]
raw_dt.to_csv('Merged_MT_DNA_variant.csv',index=False,sep=',')

```
